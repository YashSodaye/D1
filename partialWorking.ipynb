```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA

# Generate sample data
years = np.arange(2010, 2020)
division_a_budget = [100, 110, 105, 115, 120, 125, 130, 140, 135, 145]
division_b_budget = [200, 210, 215, 220, 225, 230, 235, 245, 240, 250]

data = {
    'Year': np.tile(years, 2),
    'Division': ['A'] * len(years) + ['B'] * len(years),
    'BudgetAmount': division_a_budget + division_b_budget
}

df = pd.DataFrame(data)

# Save to CSV (for demonstration purposes)
df.to_csv('sample_budget_data.csv', index=False)

# Load data
df = pd.read_csv('sample_budget_data.csv')

# Preprocess data
df['Year'] = pd.to_datetime(df['Year'], format='%Y')
df.set_index('Year', inplace=True)

# Visualize data
for division in df['Division'].unique():
    division_data = df[df['Division'] == division]
    plt.plot(division_data.index, division_data['BudgetAmount'], label=f'Division {division}')

plt.legend()
plt.xlabel('Year')
plt.ylabel('Budget Amount')
plt.title('Division-wise Budget Data')
plt.show()

# Train ARIMA model for each division
predictions = {}
for division in df['Division'].unique():
    division_data = df[df['Division'] == division]
    model = ARIMA(division_data['BudgetAmount'], order=(1, 1, 1))
    model_fit = model.fit()
    forecast = model_fit.forecast(steps=5)  # Predict next 5 years
    predictions[division] = forecast

# Output predictions
for division, forecast in predictions.items():
    print(f"Predictions for Division {division}:")
    for year, value in zip(range(2020, 2025), forecast):
        print(f"Year {year}: {value:.2f}")
```

```python
import pandas as pd
from prophet import Prophet
import matplotlib.pyplot as plt

# Load transformed data
df_long = pd.read_csv('C:/Users/yashs/Downloads/transformed_budget_data.csv')

# Preprocess data for Prophet
df_long['Year'] = pd.to_datetime(df_long['Year'], format='%Y')
df_long.reset_index(inplace=True)

# Initialize a dictionary to store predictions
predictions = {}

# Train Prophet model for each division
for division in df_long['Division'].unique():
    division_data = df_long[df_long['Division'] == division][['Year', 'BudgetAmount']]
    division_data.rename(columns={'Year': 'ds', 'BudgetAmount': 'y'}, inplace=True)
    
    model = Prophet()
    model.fit(division_data)
    
    future = model.make_future_dataframe(periods=5, freq='Y')
    forecast = model.predict(future)
    
    # Store predictions
    predictions[division] = forecast[['ds', 'yhat']].tail(5)

# Output predictions
prediction_results = {}
for division, forecast in predictions.items():
    division_predictions = {}
    for index, row in forecast.iterrows():
        year = row['ds'].year
        value = row['yhat']
        division_predictions[year] = value
    prediction_results[division] = division_predictions

# Print the prediction results
prediction_results
```
